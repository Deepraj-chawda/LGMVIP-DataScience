# Image to Pencil Sketch with Python

## Step-1:  Importing required libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Step-2:  Loading the image

```python
cat_image = cv2.imread('cat.png')
```

```python
# creating a function for showing image in notebook
def show_image(image):
    plt.figure(dpi=150)
    rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # this converts it into RGB
    plt.imshow(rgb_img)
    plt.axis('Off')
    plt.show()
    
show_image(cat_image)


#cv2.imshow('image',shinchan_image)
#cv2.waitKey(0)
```

## Step-3: Converting an image into gray_scale image

```python
gray_image = cv2.cvtColor(cat_image, cv2.COLOR_BGR2GRAY)

show_image(gray_image)

#cv2.imshow('image',gray_image)
#cv2.waitKey(0)
```

## Step-4: Inverting the image

```python
inverted_image = cv2.bitwise_not(gray_image)

show_image(inverted_image)

#cv2.imshow('image',sketch_image)
#cv2.waitKey(0)
```

## Step-5: Smoothing the image 

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (41, 41),sigmaX=0, sigmaY=0)
show_image(img_smoothing)

#cv2.imshow('image',sketch_image)
#cv2.waitKey(0)
```

### Step-6: Obtaining the final sketch

```python
sketch_image = cv2.divide(gray_image, 255 - img_smoothing, scale=256.0)
show_image(sketch_image)

#cv2.imshow('image',sketch_image)
#cv2.waitKey(0)
```
